```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Model interpolasi Lagrange
def lagrange_interpolation(f: list, xi: int, n: int) -> float:

    # Inisialisasi Hasil
    result = 0.0
    for i in range(n):

        # Compute individual terms of above formula
        term = f[i].y
        for j in range(n):
            if j != i:
                term = term * (xi - f[j].x) / (f[i].x - f[j].x)

        # Add current term to result
        result += term

    return result
```

```python
# Definisikan kelas Data
class Data:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"Data({self.x}, {self.y})"

# Daftar x dan y
x = [5, 10, 15, 20, 25, 30, 35, 40]
y = [40, 30, 25, 40, 18, 20, 22, 15]

# Buat daftar f dengan objek Data
f = [Data(xi, yi) for xi, yi in zip(x, y)]

f
```

```python
# Plot data sebelum interpolasi
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x, y)
```

```python
# Membuat garis interpolasi
x_garis = np.linspace(5, 40, 100)
y_garis = [lagrange_interpolation(f,point, 8) for point in x_garis]

plt.scatter(x, y)
plt.plot(x_garis, y_garis)

```

```python
# Deployment
x_dep = np.array ([8, 13, 18, 23, 28, 33, 38])
y_dep = [lagrange_interpolation(f,x_dep, 8)]

y_dep
```

```python
# visualisasi data
plt.figure(figsize=(10, 6))


plt.scatter (x,y,color="green", label="Titik yang diketahui")
plt.scatter(x_dep,y_dep, color="purple", label="Titik Uji", zorder=5)
plt.plot (x_garis, y_garis, label= 'Polinomial Lagrange')
plt.title('Interpolasi Polinomial Lagrange')
plt.xlabel('Tegangan, x (kg/mm²)')
plt.ylabel('Waktu patah, y (jam)')
plt.legend()
plt.grid(True)
plt.show()

```
